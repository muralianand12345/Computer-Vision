```python
#Import all necessary packages

import numpy as np
import cv2

#Draw and display image using the matplotlb
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Read the image using the image path

img = cv2.imread('/home/murali/Downloads/71955467.jpeg')
```

```python
# Display the image Usually displays in BGR

plt.imshow(img)
```

```python
# Convert the image from BGR to RGB

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
w_ratio = 0.5
h_ratio = 0.5

# the point (0,0) is init image height and width of the image
new_img = cv2.resize(img_rgb, (0,0), img_rgb, w_ratio, h_ratio)
plt.imshow(new_img)
```

```python
w_ratio = 1.5
h_ratio = 1.5

# the point (0,0) is init image height and width of the image
new_img = cv2.resize(img_rgb, (0,0), img_rgb, w_ratio, h_ratio)
plt.imshow(new_img)
```

```python
# Reading and getting the image location in Gray scale

image = cv2.imread('/home/murali/Downloads/71955467.jpeg', cv2.IMREAD_GRAYSCALE)

# Here we are applying the colormap using the applyColorMap function
colored_img = cv2.applyColorMap(image, cv2.COLORMAP_JET)
plt.imshow(colored_img)
```

```python
# The threshold function need to parameter to give the threshold value and the image itself

ret,thresh_bin = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
plt.imshow(thresh_bin,cmap='gray')
print(ret)
```

```python
# The higher the value the higher the blur image will be

blur = cv2.GaussianBlur(img,(215,215),10)
plt.imshow(blur,cmap="gray")
```

```python
avg_blur = cv2.blur(img,ksize=(21,21))
plt.imshow(avg_blur,cmap='gray')
```

```python
blur = cv2.medianBlur(img,5)
plt.imshow(blur)
```

```python
# Defining the kernal for morphology

kernel = np.ones((5, 5), np.uint8)

erosion = cv2.erode(img, kernel, iterations=1)
dilation = cv2.dilate(img, kernel, iterations=1)
opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)
```

```python
plt.imshow(img)
```

```python
plt.imshow(erosion)
```

```python
plt.imshow(dilation)
```

```python
plt.imshow(opening)
```
