```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("/content/download.jpg",0)
plt.imshow(img,cmap='gray')
```

```python
blur = cv2.GaussianBlur(img,(215,215),10)
plt.imshow(blur,cmap="gray")
```

```python
avg_blur = cv2.blur(img,ksize=(21,21))
plt.imshow(avg_blur,cmap='gray')
```
