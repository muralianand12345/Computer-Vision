```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
text = "E0120004"
font = cv2.FONT_HERSHEY_SIMPLEX
position = (50, 50)

font_scale = 2
font_color = (255, 0, 255)
font_thickness = 2
```

Reading and Writing an Image

```python
img = cv2.imread('chess.png')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
cv2.putText(img, text, position, font, font_scale, font_color, font_thickness)

plt.imshow(img)
```

```python
cv2.imwrite('chess.png',img)
```

Image Gradient

```python
image_path = 'chess.png'
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
```

```python
#Sobel filter
gradient_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
gradient_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)
cv2.putText(gradient_x, text, position, font, font_scale, font_color, font_thickness)
plt.imshow(gradient_x,cmap="gray")
```

```python
cv2.putText(gradient_y, text, position, font, font_scale, font_color, font_thickness)
plt.imshow(gradient_y,cmap="gray")
```

```python
gradient_orientation = np.arctan2(gradient_y, gradient_x) * (180 / np.pi)
cv2.putText(gradient_orientation, text, position, font, font_scale, font_color, font_thickness)
plt.imshow(gradient_orientation,cmap = 'gray')
```

```python
#prewitt filter
image_path = 'chess.png'
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
kernel_x = np.array([[-1, 0, 1],
                     [-1, 0, 1],
                     [-1, 0, 1]])

kernel_y = np.array([[-1, -1, -1],
                     [0, 0, 0],
                     [1, 1, 1]])
# Apply the Prewitt kernels to compute gradients
gradient_x = cv2.filter2D(image, cv2.CV_64F, kernel_x)
gradient_y = cv2.filter2D(image, cv2.CV_64F, kernel_y)
# Display the original image and gradient magnitude
cv2.putText(gradient_x, text, position, font, font_scale, font_color, font_thickness)
plt.imshow(gradient_x,cmap="gray")
```

```python
plt.imshow(gradient_y,cmap="gray")
cv2.putText(gradient_y, text, position, font, font_scale, font_color, font_thickness)
```

```python
gradient_magnitude = np.sqrt(gradient_x**2 + gradient_y**2)
cv2.putText(gradient_magnitude, text, position, font, font_scale, font_color, font_thickness)
plt.imshow(gradient_magnitude,cmap="gray")
```

```python
gradient_orientation = np.arctan2(gradient_y, gradient_x) * (180 / np.pi)
cv2.putText(gradient_orientation, text, position, font, font_scale, font_color, font_thickness)
plt.imshow(gradient_orientation,cmap = 'gray')
```

```python
#robert Filter
image_path = 'chess.png'
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
cv2.putText(image,text, position, font, font_scale, font_color, font_thickness)
plt.imshow(image,cmap = 'gray')
kernel_x = np.array([[1, 0],
                     [0, -1]])

kernel_y = np.array([[0, 1],
                     [-1, 0]])
gradient_x = cv2.filter2D(image, cv2.CV_64F, kernel_x)
gradient_y = cv2.filter2D(image, cv2.CV_64F, kernel_y)
```

```python
cv2.putText(gradient_x, text, position, font, font_scale, font_color, font_thickness)
plt.imshow(gradient_x,cmap="gray")
```

```python
gradient_magnitude = np.sqrt(gradient_x**2 + gradient_y**2)
plt.imshow(gradient_magnitude,cmap="gray")
```

```python
gradient_orientation = np.arctan2(gradient_y, gradient_x) * (180 / np.pi)
plt.imshow(gradient_orientation,cmap = 'gray')
```

Morphological gradient


```python
img =  cv2.imread('chess.png',cv2.IMREAD_GRAYSCALE)
plt.imshow(img,cmap = 'gray')
```

```python
se = np.ones((5,5),np.uint8)
erotion = cv2.erode(img,se,iterations = 5)
plt.imshow(erotion,cmap = 'gray')
```

```python
dilate = cv2.dilate(img,se,iterations = 2)
plt.imshow(dilate,cmap = 'gray')
```

```python
ret,thr = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
plt.imshow(thr,cmap = "gray")
```

```python
opening = cv2.morphologyEx(thr,cv2.MORPH_OPEN,se)
plt.imshow(opening,cmap = 'gray')
```

```python
closing = cv2.morphologyEx(thr,cv2.MORPH_CLOSE,se)
plt.imshow(closing,cmap = 'gray')
```

```python
gradient = cv2.morphologyEx(thr,cv2.MORPH_GRADIENT,se)
plt.imshow(gradient,cmap = 'gray')
```

```python
tophat = cv2.morphologyEx(thr,cv2.MORPH_TOPHAT,se)
plt.imshow(tophat,cmap = 'gray')
```

```python
blackhat = cv2.morphologyEx(thr,cv2.MORPH_BLACKHAT,se)
plt.imshow(blackhat,cmap = 'gray')
```
