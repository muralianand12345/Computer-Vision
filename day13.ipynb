```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
laps = cv2.imread('/home/murali/Downloads/Untitled.png')
laps = cv2.cvtColor(laps,cv2.COLOR_BGR2RGB)
plt.imshow(laps)
```

```python
lap = cv2.imread('/home/murali/Downloads/Untitled2.png')
lap = cv2.cvtColor(lap,cv2.COLOR_BGR2RGB)
plt.imshow(lap)
```

```python
result = cv2.matchTemplate(laps,lap,cv2.TM_CCOEFF_NORMED)
```

```python
threshold=0.4
locations=np.where(result>=threshold)
```

```python
thr = 0.9
location = np.where(result >= thr)
```

```python
for loc in zip(*location[::-1]):
  cv2.rectangle(laps,loc,(loc[0] + lap.shape[1],loc[1] + lap.shape[0]),(0,255,0),2)
```
