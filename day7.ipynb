```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("/home/murali/Downloads/puzzle.jpg",0)
plt.imshow(img, cmap='gray')
```

```python
def show_pic(img):
    fig = plt.figure(figsize=(15,15))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
show_pic(img)
```

```python
ret,thresh_bin = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
```

```python
plt.imshow(thresh_bin,cmap='gray')
```

```python
ret,thresh_bin_inv = cv2.threshold(img,127,255,cv2.THRESH_BINARY_INV)
plt.imshow(thresh_bin_inv,cmap='gray')
```

```python
adap = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,8)
```

```python
show_pic(adap)
```
