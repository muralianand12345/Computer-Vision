```python
import numpy as np
from numpy import genfromtxt
data = genfromtxt('bank_note_data.txt', delimiter=',', dtype=str)
```

```python
y = data[:,4]
X = data[:,0:4]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.33, random_state=42)
```

```python
y_train = y_train.astype(int)
y_test = y_test.astype(int)
```

```python
# what is random state? 
# random_state is the seed used by the random number generator.
# random_state = 42 is a common seed used for random number generator
# If we didnt use the random state, then the random number generator will generate different random numbers each time we run the code
```

```python
# variance, skewness, kurtosis, entropy, class

from sklearn.preprocessing import MinMaxScaler
scaler_object = MinMaxScaler()
scaler_object.fit(X_train)
scaled_X_train = scaler_object.transform(X_train)

# day17.ipynb

scaled_X_test = scaler_object.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(4, input_dim=4, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(scaled_X_train, y_train, epochs=50, verbose=2)
```

```python
# model.predict_classes(scaled_X_test)
predicted_probabilities = model.predict(scaled_X_test)
#predicted_classes = (predicted_probabilities > 0.5).astype(int)
predicted_classes = np.round(predicted_probabilities).astype(int)
print(predicted_classes)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
model.metrics_names
```

```python
model.evaluate(x=scaled_X_test, y=y_test)
```

```python
confusion_matrix(y_test, predicted_classes)
```

```python
classification_report(y_test, predicted_classes)
```
