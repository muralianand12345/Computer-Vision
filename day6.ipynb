### Image Thresholding

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("/home/murali/Downloads/2_62306456bfd96.jpg")
rgb_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)
```

```python
gray = cv2.imread("/home/murali/Downloads/2_62306456bfd96.jpg",0)
plt.imshow(gray,cmap = 'gray')
```

```python
ret,thresh = cv2.threshold(gray,127,255,cv2.THRESH_BINARY)
```
